# Feature Engineering


## Skew 제거

```python
from scipy.stats import skew
# float type 변수 index 추출
feature_index = data1.dtypes[data1.dtypes == 'float'].index
# train_df에 컬럼 index를 [ ]로 입력하면 해당하는 컬럼 데이터 셋 반환. apply lambda로 skew( )호출 
skew_features = data1[feature_index].apply(lambda x : skew(x))
# skew 정도가 1 이상인 컬럼들만 추출. 
skew_features_top = skew_features[skew_features > 1]
# 내림정렬
print(skew_features_top.sort_values(ascending=False))
```

```python
data1.dtypes[data1.dtypes == 'float']
```

```python
# skew 제거
data1[skew_features_top.index] = np.log1p(data1[skew_features_top.index])
data1.info()
```

##결측치 대체
#### 1. Pearson 상관관계도에 의하면 1차,2차 pulse time와 1차, 2차 purge time은 어느 정도의 선형 관계를 유지함.
#### 2. 이를 활용하여 1차, 2차 pulse time과 1차, 2차 purge time의 평균을 통해 새로운 열로 만들고 2차 pulse와 2차 purge가 없으면 1차 pulse와 purge time으로 데이터 채워 넣음

```python
# pulse time와 purge time 에 대한 feature engineering
# Create the new_purge_time column
data1['new_purge_time'] = data1['1st Purge time']

# Check if 2nd purge time column exists in the dataset
if '2nd Purge time' in data1.columns:
    # Fill new_purge_time with the average if 2nd purge time is present
    data1['new_purge_time'] = data1[['1st Purge time', '2nd Purge time']].mean(axis=1)

    # Drop the 1st purge time and 2nd purge time columns
    data1 = data1.drop(['1st Purge time', '2nd Purge time'], axis=1)

# Create the new_pulse_time column
data1['new_pulse_time'] = data1['1st Pulse time']

# Check if 2nd purge time column exists in the dataset
if '2nd Pulse time' in data1.columns:
    # Fill new_purge_time with the average if 2nd purge time is present
    data1['new_pulse_time'] = data1[['1st Pulse time', '2nd Pulse time']].mean(axis=1)

    # Drop the 1st purge time and 2nd purge time columns
    data1 = data1.drop(['1st Pulse time', '2nd Pulse time'], axis=1)

data1.info()
```

#### 상관관계도에서 보면 Pressure는 Pulse Time과 유의미한 선형적 관계를 갖고 있음.
(Gas의 flux는 pressure에 비례하기 때문으로 추측됨.)
#### 따라서, Pressure의 결측치는 Pulse Time에 비례해서 다른 데이터 값을 참조하여 결측치를 채워 넣음.

```python
# Pressure 결측치 대체
from sklearn.metrics.pairwise import pairwise_distances_argmin_min

# Find rows with missing 'Pressure' values
missing_indices = data1[data1['Pressure'].isnull()].index

# Iterate over missing indices
for index in missing_indices:
    # Get the new_pulse_time value of the current row
    current_new_pulse_time = data1.loc[index, 'new_pulse_time']
    
    # Exclude rows with NaN values in 'new_pulse_time' column
    valid_indices = data1[data1['new_pulse_time'].notnull()].index
    
    # Calculate pairwise distances between current new_pulse_time value and valid rows
    distances = pairwise_distances_argmin_min(data1.loc[valid_indices, 'new_pulse_time'].values.reshape(-1, 1),
                                              np.array([[current_new_pulse_time]]))
    
    # Find the index of the row with the most similar new_pulse_time value
    similar_index = valid_indices[distances[0][0]]
    
    # Get the pressure value from the row with the similar new_pulse_time value
    similar_pressure = data1.loc[similar_index, 'Pressure']
    
    # Replace the missing 'Pressure' value with the similar pressure value
    data1.loc[index, 'Pressure'] = similar_pressure

data1.info()
```

### Atomosphere 종류에 따라 Pressure_a와 Time이 결정됨. 




#### 이를 이용하여 Pressure_a, Time 결측치 채워 넣음.





```python
# Create a dictionary mapping Atomosphere values to corresponding Pressure_a values
pressure_mapping = {'Vacuum': 0, 'air': 760, 'N2': 760, 'O2': 760}

# Fill missing values in Pressure_a based on Atomosphere values
data1['Pressure_a'] = data1.apply(lambda row: row['Pressure_a'] if not pd.isnull(row['Pressure_a']) else pressure_mapping.get(row['Atomosphere']), axis=1)

data1.info()
```

```python
# Calculate the average time for each Atomosphere value
average_time = data1.groupby("Atomosphere")["Time"].median()

# Fill in the missing values based on the corresponding average time
data1["Time"] = data1.apply(lambda row: average_time[row["Atomosphere"]] if pd.isnull(row["Time"]) else row["Time"], axis=1)

data1.info()
```

## 평가 지표 확인(10-fold cross validation)

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error, r2_score
import seaborn as sns
# Create a list to store the results for each metric
results = []

# Define the models
models = [
    RandomForestRegressor(random_state=42, max_depth=None, max_features='log2', min_samples_leaf=1, min_samples_split=2, n_estimators=200),
    ExtraTreesRegressor(max_depth=None, max_features='sqrt', min_samples_leaf=1, min_samples_split=2, n_estimators=100, random_state=42),
    AdaBoostRegressor(base_estimator=ExtraTreeRegressor(), learning_rate=0.01, loss='square', n_estimators=200, random_state=42),
    XGBRegressor(random_state=42, gamma=0.1, learning_rate=0.1, max_depth=7, min_child_weight=1, n_estimators=300),
    CatBoostRegressor(depth=9, iterations=200, l2_leaf_reg=1, learning_rate=0.1, random_strength=0.4)
]

model_labels = ['Random Forest', 'Extra Trees', 'AdaBoost', 'XGBoost', 'CatBoost']

# Define the scoring metrics
scoring = {
    'r2': make_scorer(r2_score),
    'rmse': make_scorer(mean_squared_error, squared=False),
    'mae': make_scorer(mean_absolute_error)
}
# Replace problematic characters in feature names
X.columns = X.columns.str.replace('[', '_').str.replace(']', '_').str.replace('<', '_')

# Perform 10-fold cross-validation for each model and collect the results
for model, label in zip(models, model_labels):
    cv_results = cross_validate(model, X, y, cv=10, scoring=scoring)
    
    # Add the scores to the results list
    results.append({'Model': label, 'Metric': 'R2', 'Score': cv_results['test_r2']})
    results.append({'Model': label, 'Metric': 'RMSE', 'Score': cv_results['test_rmse']})
    results.append({'Model': label, 'Metric': 'MAE', 'Score': cv_results['test_mae']})
```

```python
# Create a DataFrame from the results
results_df = pd.DataFrame(results)
# Create subplots for each metric
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

# Plot box plots for each metric
for i, metric in enumerate(['R2', 'RMSE', 'MAE']):
    # Filter the results for the specific metric
    metric_df = results_df[results_df['Metric'] == metric]

    # Convert the 'Score' column values to numeric
    metric_df['Score'] = metric_df['Score'].apply(lambda x: x[0])

    # Plot the box plot
    sns.boxplot(data=metric_df, x='Model', y='Score', ax=axes[i], linewidth=2.5)
    axes[i].set_ylabel(metric)
    axes[i].set_xlabel('Model')
    axes[i].set_title(metric)

# Add a title
fig.suptitle('Cross-Validation Results')

# Rotate x-axis tick labels
plt.xticks(rotation=45)

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()

```
